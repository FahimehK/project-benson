```python
import pandas as pd
```

```python
# read the combined txt file into dataframe
mta = pd.read_csv("/Users/monkeyjiashuo/Documents/GitHub/sf19_ds15/curriculum/project-01/Project-Benson/Data Files/combined_file.txt")
```

```python
# check the # of rows&columns and data types in dataframe
mta.info()
```

```python
# randomly sample 50 instances to get sense of dataframe
mta.sample(50, random_state = 22)
```

```python
# all distinct NY subway station names
mta.STATION.unique()
```

```python
# count frequencies of each station
station_counts = mta['STATION'].value_counts()
station_counts
```

```python
# get index of those value_counts greater than 10000 - returns the station name as index
gt_10000 = station_counts[station_counts > 10000].index
gt_10000
```

```python
# create a new dataframe including only stations with frequency greater than 10000
most_freq_station = mta[mta['STATION'].isin(gt_10000)]
most_freq_station
```

```python
most_freq_station['STATION'].unique()
```

```python
# create interactive barplot to see the frequency distribution of most_freq_station
%matplotlib notebook
most_freq_station.STATION.value_counts().plot('barh')
```
